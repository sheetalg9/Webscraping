## Data Collection

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests 
from bs4 import BeautifulSoup

sns.set_style('whitegrid')
plt.style.use('bmh')

import warnings
warnings.filterwarnings('ignore')

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

```

```python
url = 'https://www.imdb.com/search/title/?groups=top_1000&ref_=adv_prv'
page = requests.get(url)
pagecontent = page.text
```

```python
soup = BeautifulSoup(pagecontent)
soup
```

```python
x = soup.find('div',attrs={'class':"lister-item-content"})
x
```

```python
import time
movie_names=[]
release_year=[]
movie_ranking=[]
movie_runtime =[]
movie_genre=[]
movie_rating=[]
metascore=[]
director=[]
gross=[]
votes=[]
pgno=[]

for i in range(1,602,50):
    start_time =time.time()
    url ="https://www.imdb.com/search/title/?groups=top_1000&start={}&ref_=adv_nxt".format(i)
    pagecontent = requests.get(url).text
    soup = BeautifulSoup(pagecontent)
    
            
    for x in soup.find_all('div',attrs={'class':"lister-item mode-advanced"}):
        movie_name = x.h3.a.text.strip()
        if movie_name is None:
            movie_names.append(np.NaN)
        else:
            movie_names.append(movie_name)
            
        year = x.find('span',attrs = {'class':"lister-item-year text-muted unbold"})
        if year is None:
            release_year.append(np.NaN)
        else:
            release_year.append(year.text[1:5].strip())
           
        ranking = x.find('span',attrs={'class':"lister-item-index unbold text-primary"})
        if ranking is None:
            movie_ranking.append(np.NaN)
        else:
            movie_ranking.append(ranking.text)
           
       
        runtime = x.find('span',attrs={'class':'runtime'})
        if runtime is None:
            movie_runtime.append(np.NaN)
        else:
            movie_runtime.append(runtime.text.strip())
            
        genre = x.find('span',attrs={'class':'genre'})
        if genre is None:
            movie_genre.append(np.NaN)
        else:
            movie_genre.append(genre.text.strip())
        
        
        rating = x.find('div',attrs={'class':'inline-block ratings-imdb-rating'})
        if rating is None:
            movie_rating.append(np.NaN)
        else:
            movie_rating.append(rating.text.strip())
        
        
        meta = x.find('div',attrs = {'class':"inline-block ratings-metascore"})
        if meta is None:
            metascore.append(np.NaN)
        else:
            metascore.append(meta.text.strip())
        
        direct = x.find('p',attrs={'class':""})
        if direct is None:
            director.append(np.NaN)
        else:
            director.append(direct.text.strip())
        
        vote = x.find('span',attrs = {'name':"nv"})
        if vote is None:
            votes.append(np.NaN)
        else:
            votes.append(vote.text)
            
            
        grs = x.find('p',attrs={'class':'sort-num_votes-visible'})
        if grs is None:
            gross.append(np.NaN)
        else:
            gross.append(grs.text)
        
        pgno.append(i)
    end_time=time.time()
    print('Page {} completed in {} seconds'. format(i, end_time-start_time))
```

```python
imdb = pd.DataFrame({'movie_names':movie_names,
                    'release_year':release_year,
                    'movie_ranking':movie_ranking,
                    'movie_runtime':movie_runtime,
                    'movie_genre':movie_genre,
                    'movie_rating':movie_rating,
                    'metascore':metascore,
                    'director':director,
                    'votes':votes,
                    'Gross':gross})
                     
```
